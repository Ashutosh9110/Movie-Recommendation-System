```python
import numpy as np
import pandas as pd
import difflib #It is used to find the close match of the input(movie name here) submitted by the user. It will compare
# it with the dataset (moviedata.csv) that we have. Eg: Spelling mistake
from sklearn.feature_extraction.text import TfidfVectorizer # It is used to convert textual data into numerical values. 
# We will be converting textual data into meaningful number which are Feature Vector
from sklearn.metrics.pairwise import cosine_similarity 
# This will help in finding the similarity confidence score of all the movies which will then help in recommendations.
```

# Data Collection and Pre-Processing

```python
#  loading the data from the csv file to pandas dataframe
```

```python
movies_data = pd.read_csv("C:\\Users\\veni_\\Desktop\\Movie Recommendation System\\moviedata.csv")
```

```python
# printing the first 5 rows of the dataframe
movies_data.head()
```

```python
# number of rows and columns in the data frame

movies_data.shape
```

```python
# selecting the relevant features for recommendation

selected_features = ['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
# replacing the null valuess with null string

for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
# combining all the 5 selected features

combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
print(combined_features)
```

```python
# converting the text data to feature vectors

vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features) # Storing (fitting and transforming 
# all the numerical values.
```

```python
print(feature_vectors) # This is how combined_features data would look like after gettng converted into numerical Data/
```

# Using Cosine Similarity

```python
# getting the similarity scores using cosine similarity

similarity = cosine_similarity(feature_vectors) # the above converted data will now be fed to cosine_similarity to find 
# similar values
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

# Getting the movie name from the user

```python
movie_name = input('Enter your favourite movie name : ')
```

```python
# creating a list with all the movie names given in the dataset

list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
# finding the close match for the movie name given by the user

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
# finding the index of the movie with title

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
# getting a list of similar movies

similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# sorting the movies based on their similarity score

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)  
print(sorted_similar_movies)

# Reverse is used to sort the values in descending order
# x:x[1] represents the similarity_score i.e. first x is for the index number and the second x is for the score
```

```python
# print the name of similar movies based on the index

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```

# Movie Recommendation Sytem

```python
movie_name = input(' Enter your favourite movie name : ')

list_of_all_titles = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

close_match = find_close_match[0]

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
